```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x, n):
    fx = np.pi/2.
    for i in range(1, n+1):
        if i%2 != 0:
            fx += 2*np.sin(n*x)/n
    return fx
```

```python
x = np.linspace(-np.pi, np.pi, 1001)
nlist = [11, 21, 31]
plt.figure(figsize = (16,10))
for n in nlist:
    fx = func(x, n)
    plt.plot(x, fx, label = "n = "+str(n))
plt.legend(fontsize = 18, loc = "upper right")
plt.xlim(x[0],x[-1])
plt.show()
```
