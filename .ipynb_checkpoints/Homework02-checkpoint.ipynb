```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def func(x, n):
    fx = np.pi/2.
    for i in range(1, n+1):
        if i%2 != 0:
            fx += 2*np.sin(i*x)/i
    return fx
```

```python
x = np.linspace(-np.pi, np.pi, 1001)
nlist = [11, 21, 31]
plt.figure(figsize = (16,10))
for n in nlist:
    fx = func(x, n)
    plt.plot(x, fx, label = "n = "+str(n))
plt.legend(fontsize = 18, loc = "upper left")
plt.xlim(x[0],x[-1])
plt.show()
```

```python
def fib2(n):
    if n == 0: return 0
    elif n == 1: return 1
    else: return fib2(n-1)+fib2(n-2)
```

```python
n_list1 = np.array([1, 5, 10, 20, 25, 45])   # 45 is the best I can get.
time1 = np.zeros(n_list1.size)

for i in range(n_list1.size):
    start_time = time.time()
    f = fib2(n_list1[i])
    time1[i] = time.time() - start_time
    print("n = ", n_list1[i])
    print("--- %s seconds ---" % time1[i])
```

```python
def fib4(n):
    f0 = 0
    f1 = 1
    if n == 0:
        return f0
    elif n == 1: 
        return f1
    else:
        for i in range(2,n+1):
            result = f0+f1
            f0 = f1
            f1 = result
        return result
```

```python
n_list2 = np.array([10**i for i in range(8)])
time2 = np.zeros(n_list2.size)

for i in range(n_list2.size):
    start_time = time.time()
    f = fib4(n_list2[i])
    time2[i] = time.time() - start_time
    print("n = ", n_list2[i])
    print("--- %s seconds ---" % time2[i])
```

```python
def arrpow(arr, n):
    yarr=arr
    if n<1:
        print('n needs to be larger than 1')
    if n==1:
        return arr
    yarr = arrpow(arr, n//2)
    yarr = [[yarr[0][0]*yarr[0][0]+yarr[0][1]*yarr[1][0],yarr[0][0]*yarr[0][1]+yarr[0][1]*yarr[1][1]],
            [yarr[1][0]*yarr[0][0]+yarr[1][1]*yarr[1][0],yarr[1][0]*yarr[0][1]+yarr[1][1]*yarr[1][1]]]
    if n%2:
        yarr=[[yarr[0][0]*arr[0][0]+yarr[0][1]*arr[1][0],yarr[0][0]*arr[0][1]+yarr[0][1]*arr[1][1]],
            [yarr[1][0]*arr[0][0]+yarr[1][1]*arr[1][0],yarr[1][0]*arr[0][1]+yarr[1][1]*arr[1][1]]]
    return yarr

def fib5(n):
    if n == 0: 
        return 0
    elif n== 1:
        return 1
    else:
        arr= [[1,1],[1,0]]
        f=arrpow(arr,n-1)[0][0]
    return f
```

```python
n_list3 = np.array([10**i for i in range(9)])
time3 = np.zeros(n_list3.size)

for i in range(n_list3.size):
    start_time = time.time()
    f = fib5(n_list3[i])
    time3[i] = time.time() - start_time
    print("n = ", n_list3[i])
    print("--- %s seconds ---" % time3[i])
```

```python
print("Accuracy testing: fib(10) = ")
n = 10
print("Recursion method: ", fib2(n))
print("Tabulation method: ", fib4(n))
print("matrix method: ", fib5(n))
```

```python
plt.figure(figsize = (10,8))
plt.plot(n_list1, time1, label = "Recursion")
plt.plot(n_list2, time2, label = "Tabulation")
plt.plot(n_list3, time3, label = "Matrix")
plt.legend(fontsize = 18, loc = "lower right")
plt.semilogx()
plt.semilogy()
plt.xlabel("n", fontsize = 18)
plt.ylabel("time", fontsize = 18)
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.plot(np.log10(n_list1), np.log10(time1), label = "Recursion")
plt.plot(np.log10(n_list2), np.log10(time2), label = "Tabulation")
plt.plot(np.log10(n_list3), np.log10(time3), label = "Matrix")
plt.legend(fontsize = 18, loc = "lower right")
plt.xlabel("log10(n)", fontsize = 18)
plt.ylabel("log10(time)", fontsize = 18)
plt.show()
```

The complexity of Recursion method is $O(2^n)$. It still grow exponentially in a log-log plot. 

Tabulation method is $O(n)$. It's almost a straight line.

And matrix method is $log_2(n)$.

For small n, matrix method is slower because it has to initialize the matrix. But for large n, It becomes much faster
